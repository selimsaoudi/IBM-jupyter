## Exemples d'expressions arithmétiques

Quelques exemples simples de calculs que l'on peut faire en Python :

- `(2 + 3) * 4` → addition puis multiplication
- `10 ** 2` → puissance (10 au carré)
- `15 / 4` → division classique
- `15 // 4` → division entière (quotient)
- `15 % 4` → reste de la division euclidienne

Nous allons maintenant écrire du code Python pour réaliser des calculs similaires.

```
# Exercice 8 : addition et multiplication de nombres

# Nous définissons deux nombres
x = 24
y = 5

# Calcul d'une expression arithmétique
resultat = (2 * x) + y

print("x =", x)
print("y =", y)
print("Résultat de (2 * x) + y =", resultat)
```

```
# Exercice 9 : conversion de minutes en heures

# On souhaite convertir un nombre de minutes en heures et minutes restantes

minutes = 135  # exemple : 135 minutes

heures = minutes // 60
minutes_restantes = minutes % 60

print(f"Durée totale : {minutes} minutes")
print(f"Soit {heures} heure(s) et {minutes_restantes} minute(s)")
```
